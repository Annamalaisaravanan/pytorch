```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import numpy as np
import pandas as pd
```

```python
my_data = pd.read_csv('E:/Dataset/P14-Simple-Linear-Regression/Simple_Linear_Regression/Salary_Data.csv')
my_data.head()
```

```python
x_data = my_data.iloc[:,0].values
y_data = my_data.iloc[:,1].values
```

```python
X = torch.Tensor(x_data.reshape(-1,1).astype(np.float32))
```

```python
X.shape
```

```python
y = torch.Tensor(y_data.astype(np.float32))
y = y.view(y.shape[0],1)
```

```python
y.shape
```

```python
class linear_reg(nn.Module):
            def __init__(self):
                     super(linear_reg,self).__init__()
                     self.linear = nn.Linear(1,1)

            def forward(self,x):
                       y_pred = self.linear(x)
                       return y_pred
        
```

```python
model = linear_reg()
criteria = nn.MSELoss()
optimizer =torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
for i in range(1,1000):
        #forward
        y_pred = model.forward(X)
        
        loss = criteria(y_pred,y)
        
        loss.backward()
        
        optimizer.step()
        
        optimizer.zero_grad()
        
        print(f'epochs {i} and loss {loss.item()}')
        
```

```python
import matplotlib.pyplot as plt
predicted = model(X).detach().numpy()
plt.plot(x_data,y_data,'ro')
plt.plot(x_data,predicted,'b')
plt.show()
```
