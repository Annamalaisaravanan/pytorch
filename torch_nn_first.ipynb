```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import numpy as np
import pandas as pd
```

```python
my_data = pd.read_csv("E:/Dataset/P14-Logistic-Regression/Logistic_Regression/Social_Network_Ads.csv")
my_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
my_data['Gender'] = label.fit_transform(my_data['Gender'])
my_data.head()
```

```python
x_data = my_data.iloc[:,1:-1].values
y_data = my_data.iloc[:,-1].values
```

```python
y = torch.Tensor(y_data.astype(np.float32))
y = y.view(y.shape[0],1)
X = torch.Tensor(x_data.reshape(-1,1).astype(np.float32))
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
new_x = scale.fit_transform(x_data)
```

```python
NEWX = torch.Tensor(torch.from_numpy(new_x.astype(np.float32)))
```

```python
class neural(nn.Module):
                        def __init__(self,inp,hid,out):
                                super(neural,self).__init__()
                                self.l1 = nn.Linear(inp,hid)
                                self.relu = torch.nn.ReLU()
                                self.l2 = nn.Linear(hid,out)
                    
                        def forward(self,x):
                                    out = self.l1(x)
                                    out = self.relu(out)
                                    out = self.l2(out)
                                    out = torch.sigmoid(out)
                                    return out
```

```python
model = neural(3,6,1)

criteria = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
torch_prob, torch_label =torch.max(model.forward(NEWX),1)
torch_label = torch_label.view(-1,1)
torch_label.shape
```

```python
device = torch.device('cpu')
```

```python
y = y.to(device)
torch_label.shape,y.shape
```

```python
model.forward(NEWX)
```

```python
for i in range(1,501):
                out = model.forward(NEWX).round()
                loss = criteria(out,y)
                loss.backward()
                optimizer.step()
                optimizer.zero_grad()
                if i%10==0:
                     print(f"epochs {i} and loss {loss.item()}")
with torch.no_grad():
        y_predicted = model(NEWX)
        y_predicted_cls = y_predicted.round()
        acc = y_predicted_cls.eq(y).sum()/len(y)
        print(f"the accuracy is {acc}")
                
```
