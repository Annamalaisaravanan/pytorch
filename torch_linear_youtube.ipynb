```python
import torch
from torch.autograd import Variable
import numpy as np
import pandas as pd
import sklearn
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import torch.nn as nn
```

```python
x_data,y_data = datasets.make_regression(n_samples = 100, n_features=1,noise=20,random_state=1)
```

```python
X = torch.from_numpy(x_data.astype(np.float32))
y = torch.from_numpy(y_data.astype(np.float32))
```

```python
x_data.shape,y_data.shape
```

```python
y = y.view(y.shape[0],1)
```

```python
X.shape,y.shape
```

```python
n_samples,n_features = X.shape
```

```python
class linearmodel(nn.Module):
                     def __init__(self):
                           super(linearmodel,self).__init__()
                           self.linear = nn.Linear(1,1)

                     def forward(self,x):
                            pred = self.linear(x)
                            return pred
```

```python
models = linearmodel()
```

```python
criteria = nn.MSELoss()
optimizer = torch.optim.SGD(models.parameters(),lr=0.01)
```

```python
for i in range(1,700):
        
        y_pred = models.forward(X)
        loss = criteria(y_pred,y)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        print(f'epochs {i} and loss {loss.item()}')
```

```python
predicted = models(X).detach().numpy()
```

```python
plt.plot(x_data,y_data,'ro')
plt.plot(x_data,predicted,'b')
plt.show()
```
